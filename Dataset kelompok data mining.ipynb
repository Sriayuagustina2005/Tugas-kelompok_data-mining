```python
#mengitung dan menampilkan mean, median, mode, standar deviasi, variansi, skewness, dan kurtosis untuk kolom numerik.
import pandas as pd
import numpy as np

# Memilih hanya kolom numerik
df_numeric = df.select_dtypes(include=[np.number])

# Menghitung statistik dasar untuk setiap kolom numerik
mean_values = df_numeric.mean()                      # Mean
median_values = df_numeric.median()                  # Median
mode_values = df_numeric.mode().iloc[0]              # Mode (hanya mengambil mode pertama jika ada beberapa)
std_dev_values = df_numeric.std()                    # Standar Deviasi
variance_values = df_numeric.var()                   # Variansi
skewness_values = df_numeric.skew()                  # Skewness
kurtosis_values = df_numeric.kurt()                  # kurtosis

# Menampilkan hasil
print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
print("\nStandar Deviasi:\n", std_dev_values)
print("\nVariansi:\n", variance_values)
print("\nSkewness:\n", skewness_values)
print("\nKurtosis:\n", kurtosis_values)

#dtype: float64 adalah tipe data untuk angka desimal yang sangat akurat. 
#Dengan 64 bit, tipe ini dapat menyimpan angka desimal dengan tingkat presisi yang tinggi, memungkinkan untuk perhitungan yang sangat tepat dan akurat.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pilih hanya kolom numerik
numeric_cols = df.select_dtypes(include='number')

# Membuat histogram untuk setiap kolom numerik
for column in numeric_cols.columns:
    plt.figure(figsize=(10, 6))
    plt.hist(numeric_cols[column].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan df adalah DataFrame
# Memilih hanya kolom numerik
df_numerik = df.select_dtypes(include='number')

# Menghitung matriks korelasi
correlation_matrix = df_numerik.corr()

# Menemukan pasangan variabel dengan korelasi tertinggi (absolut)
correlation_pairs = correlation_matrix.unstack().sort_values(ascending=False)
# Menghapus korelasi diri sendiri
correlation_pairs = correlation_pairs[correlation_pairs < 1]
# Mengambil pasangan dengan korelasi tertinggi
highest_correlation = correlation_pairs.idxmax()

# Membuat scatter plot dari pasangan variabel dengan korelasi tertinggi
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_numerik[highest_correlation[0]], y=df_numerik[highest_correlation[1]])
plt.title(f"Scatter Plot {highest_correlation[0]} vs {highest_correlation[1]}")
plt.xlabel(highest_correlation[0])
plt.ylabel(highest_correlation[1])
plt.grid(True)
plt.show()
```
